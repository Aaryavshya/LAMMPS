```python
import re
import glob
import numpy as np
import matplotlib.pyplot as plt

glob.glob("*.dat")
```

```python
name = str(input("Enter the name of the file"))
timestep = float(input("Enter the timestep of simulation"))
extracted_data_x = []
extracted_data_y = []
x = []
y = []
for file in glob.glob(f"{name}.dat"):
    with open(file) as f:
        for line in f:
            if "#" not in line:
                extracted_data_x.append(line.replace("\n","").split(' ')[0])
                extracted_data_y.append(line.replace("\n","").split(' ')[1])
```

```python
for i in extracted_data_x:
    xval = float(i)*timestep
    x.append(xval)

for i in extracted_data_y:
    yval = float(i)
    y.append(yval)
```

```python
x,y
```

```python
plt.plot(x,y,'-ob')
plt.xlabel("time")
plt.ylabel(f"{name}")
plt.title(f"{name} Plot")
plt.show()
```
